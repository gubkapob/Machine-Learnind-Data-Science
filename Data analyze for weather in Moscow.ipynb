```python
data['Data'] = pd.to_datetime(data['Местное время в Москве (ВДНХ)'], dayfirst=True)
data.head(5)
```

```python
x = data['Data']
y = data['T']

plt.figure(figsize=(20,5))
plt.plot(x, y, label='Temp')
plt.legend()
```
