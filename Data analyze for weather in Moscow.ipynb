```python
x = data_short['Data']
y = data_short['T']

plt.figure(figsize=(20,5))
plt.scatter(x, y, label = 'Data', s=2, c='blue')
plt.legend()
```

```python
#ВЫбор диапазона данных (приближение участка с самой низкой температурой). Способ №1.
cond2 = (data['Data'] > '2017.01.01') & (data['Data'] < '2017.01.13')

data_short2 = data[cond2]

plt.figure(figsize=(20,5))
plt.scatter(data_short2['Data'], data_short2['T'], s=2, c='blue')
```

```python
#ВЫбор диапазона данных (приближение участка с самой низкой температурой). Способ №2.
#cond2 = (data['Data'] > '2017.01.01') & (data['Data'] < '2017.01.13')
cond2 = data['Data'].between('2017.01.01', '2017.01.13')

data_short2 = data[cond2]

plt.figure(figsize=(20,5))
plt.scatter(data_short2['Data'], data_short2['T'], s=2)
```

Готовим датасет для модели. Разделяем данные на тренировочный и тестовый датасеты.

```python
data_train = data[data['Data'] < '2020.01.01']
data_test = data[data['Data'] >= '2020.01.01']
```

```python
data_train.shape, data_test.shape
```

```python
plt.figure(figsize=(20,5))
plt.scatter(data_train['Data'], data_train['T'], label = 'data_train', s=2)
plt.scatter(data_test['Data'], data_test['T'], label = 'data_test', s=2)
plt.legend()
```

'Feature engineering' - работа с признаками. Генерируем новые признаки - фичи и/или преобразуем имеющиеся величины для прогноза

```python
#Создаем новый признак
data['dayofyear'] = data['Data'].dt.dayofyear

#Заново переразбиваем датасет на train-test чтобы изменения применились 
data_train = data[data['Data'] < '2020.01.01']
data_test = data[data['Data'] >= '2020.01.01']
```

```python
plt.figure(figsize=(20,5))
plt.scatter(data_train['dayofyear'], data_train['T'], label = 'data_train', s=2)
plt.scatter(data_test['dayofyear'], data_test['T'], label = 'data_test', s=2)
plt.legend()
```
